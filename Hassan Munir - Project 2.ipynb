### Storytelling Data Visualization on Exchange Rates
there are two types of data visualization:
 * Exploratory data visualization: we create graphs for ourselves to better understand and explore data.
 * Explanatory data visualization: we create graphs for others to inform, make a point, or tell a story.

Throughout the Project, we focused on explanatory data visualization and learned the following:

* How to use information design principles (familiarity and maximizing the data-ink ratio) to create better graphs for an audience.
* About the elements of a story and how to create storytelling data visualizations using Matplotlib.
* How to guide the audience's attention with pre-attentive attributes.
* How to use Matplotlib built-in styles — with a case study on the FiveThirtyEight style.









```python
import pandas as pd
import warnings

warnings.filterwarnings("ignore")

exchange_rates=pd.read_csv("euro-daily-hist_1999_2022.csv")
```

```python
exchange_rates.rename(columns={'[US dollar ]':'US_dollar','Period\\Unit:':'Time'},inplace=True)

exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])

exchange_rates.sort_values('Time', inplace=True)

exchange_rates.head()
```

```python
euro_to_dollar = exchange_rates[['Time','US_dollar']].copy()
euro_to_dollar['US_dollar'].value_counts() #62 '-' characters(may be holidays)

```

```python
euro_to_dollar=euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

# ROLLING MEAN



```python
import matplotlib.pyplot as plt

plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.show()
```

If we look at the line's shape, we see many small wiggles — rather than seeing a smooth line. The wiggles, however, have meaning: : they are the visual representation of the daily variation in the exchange rate. The rate goes up and down, up and down again, day to day. The rate only shows clear upward or downward trends in the longer run (months or years).

Depending on our goals, we may not want to show that daily variation on our graph. If we want to hide it and show only the long-term trends, we can use the rolling mean (also known as the moving average).

```python
values = pd.DataFrame()
values["daily_values"] = pd.Series(range(1,20,2))
values
```

```python
values["rolling_mean_2"] = values["daily_values"].rolling(2).mean()
values
```

```python
values["rolling_mean_3"] = values["daily_values"].rolling(3).mean()
values["rolling_mean_5"] = values["daily_values"].rolling(5).mean()
values
```

```python
plt.figure(figsize=(16,13))

plt.subplot(3,2,1)
plt.plot(euro_to_dollar["Time"],euro_to_dollar["US_dollar"])
plt.title('Original Values',weight='bold')
plt.ylim(0.8,1.7)

for i, rolling_mean in zip([2,3,4,5,6,],[7,30,50,100,365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'],euro_to_dollar['US_dollar'].rolling(rolling_mean).mean())
    plt.title('Rolling Window:' + str(rolling_mean),weight='bold')
    plt.ylim(0.8,1.7)

plt.tight_layout() # Auto_adjust the padding between the subplots
plt.show()
```

# STORYTELLING DATA VISUALIZATION
Financial Crisis 2007_2008

```python
financial_crisis = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >=2006) & (euro_to_dollar['Time'].dt.year <=2009)]

financial_crisis_7_8 = euro_to_dollar.copy()[(euro_to_dollar.Time.dt.year >=2007) & (euro_to_dollar.Time.dt.year <=2008)]
```

```python
import matplotlib.style as style
style.use('fivethirtyeight')

### Adding the plot
fig,ax=plt.subplots(figsize=(8,3))
ax.plot(financial_crisis['Time'],financial_crisis['rolling_mean'],linewidth=2,color='#77c3ec')

### Highlighting the 2007_2008 period
ax.plot(financial_crisis_7_8['Time'],financial_crisis_7_8['rolling_mean'],linewidth=3,color='#00008B')

# print(ax.set_xticklabels(["2006","2007","2008","2009","2010"],alpha=0.3))

ax.set_xticklabels([])
x = 0.02
for year in ["2006","2007","2008","2009","2010"]:
    ax.text(x,-0.08,year,alpha=0.5,fontsize=11,transform=plt.gca().transAxes)
    x+= 0.22888
    
# print(plt.yticks())
ax.set_yticklabels([])

y = 0.07
for rate in ['1.2','1,3','1.4','1.5']:
    ax.text(-.04,y,rate,alpha=0.5,fontsize=11,transform=plt.gca().transAxes)
    y += 0.2333

## Adding a title and a subtitle
ax.text(-.05,1.2,"Euro_usd rate peaked at 1.59 during 2007_2008's financial crisis",weight='bold',transform=plt.gca().transAxes)
ax.text(-.05,1.1,"Euro_usd exchanged rate between 2006_2010 ",size=12,transform=plt.gca().transAxes)

### Adiing a signature
ax.axvspan(xmin=pd.to_datetime("2008-4-1"),xmax=pd.to_datetime("2008-9-1"),ymin=0.09,alpha=0.3,color="red")

plt.show()
```

This graph highlights the 2007–2008 financial crisis period by following the Euro to US dollar exchange rate from 2006 to 2010. As the rate rises, the Dark blue line represents the strengthening of the Euro against the US dollar, which began in 2006. The crisis's high point, represented by the shaded area, is reached around 1.59, after which there follows a dramatic decline. As the line becomes light blue once more after the crisis, it indicates that the rate has stabilized and is starting to rebound, indicating some economic recovery or stability by 2010.

 # Covid19

```python
corona_crisis_20=euro_to_dollar.loc[(euro_to_dollar['Time']>="2020-01-01")&(euro_to_dollar['Time']<="2020-12-31")]

corona_crisis=euro_to_dollar.loc[(euro_to_dollar['Time']>="2016-01-01")&(euro_to_dollar['Time']<="2019-12-31")]
```

```python
import matplotlib.style as style
style.use('fivethirtyeight')

### Adding the plot
fig,ax=plt.subplots(figsize=(9,3))
ax.plot(corona_crisis['Time'],corona_crisis['rolling_mean'],linewidth=2,color='#77c3ec')

### Highlighting the 2020 period
ax.plot(corona_crisis_20['Time'],corona_crisis_20['rolling_mean'],linewidth=3,color='#00008B')

# print(ax.set_xticklabels(["2016","2017","2018","2019","2020","2021"],alpha=0.5))
ax.set_xticklabels([])
x = 0.02
for year in ["2016","2017","2018","2019","2020"]:
    ax.text(x,-0.08,year,alpha=0.5,fontsize=11,transform=plt.gca().transAxes)
    x+=0.22888
    
 # print(plt.yticks())
ax.set_yticklabels([])

y = 0.02
for rate in ['1.05','1,10','1.15','1.20']:
    ax.text(-.04,y,rate,alpha=0.5,fontsize=11,transform=plt.gca().transAxes)
    y += 0.248

## Adding a title and a subtitle
ax.text(-.05,1.2,"Euro_usd rate peaked at 1.22 during 2020's COVID19 crisis",weight='bold',transform=plt.gca().transAxes)
ax.text(-.05,1.1,"Euro_usd exchanged rate between 2016 and 2020 ",size=12,transform=plt.gca().transAxes)

plt.show()
   



```

The fluctuation in the US dollar to euro exchange rate between 2016 and 2020 is shown in this graph. Up until the end of 2019, it indicates a general downward trend in the euro's value relative to the US dollar. The shift in the line colour from Light blue to Dark blue, however, indicates a significant rise in 2020 during the COVID-19 crisis, with the rate peaking at 1.22. This peak most likely represents the market's response to the economic effects of the pandemic, as investor mood, economic policy, and world events may all have an impact on currency prices.

# The Three US Presidencies Example

```python
bush_obama_trump = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2001) & (euro_to_dollar['Time'].dt.year < 2021)]

bush = bush_obama_trump.copy()[bush_obama_trump['Time'].dt.year < 2009]

obama = bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year >=2009) & (bush_obama_trump['Time'].dt.year < 2017)]

trump = bush_obama_trump.copy()[(bush_obama_trump['Time'].dt.year >= 2017) & (bush_obama_trump['Time'].dt.year < 2021)]

```

```python
### Adding the FiveThirtyEight Style
style.use('fivethirtyeight')

### Adding the subplots
plt.figure(figsize=(14,8))

# Pattern 1
ax1 = plt.subplot(3,3,1) #row, col,index
ax2 = plt.subplot(3,3,2)
ax3 = plt.subplot(3,3,3)

# Pattern 2
ax4 = plt.subplot(3,1,2) #row, col,index

axes = [ax1, ax2, ax3, ax4]

### Change to all the subplots

for ax in axes:
    ax.set_ylim (0.8, 1.7)
    ax.set_yticks([1.0,1.2,1.4,1.6])
    ax.set_yticklabels(['1.0', '1.2', '1.4', '1.6$'], alpha = 0.4)
    
### Ax1 : Bush
ax1.plot(bush['Time'], bush['rolling_mean'], color= '#6A0DAD')
ax1.set_xticklabels(['','2001','','2003','','2005','','2007','','2009'], alpha = 0.3, size = 12)
ax1.text(0.099, 2.45, 'BUSH', fontsize = 18, weight='bold', color='#6A0DAD', transform = plt.gca().transAxes)
ax1.text(0.08, 2.34, '(2001-2009)', weight='bold', alpha=0.3, transform = plt.gca().transAxes)

### Ax2 : Obama
ax2.plot(obama['Time'], obama['rolling_mean'], color= '#ffa500')
ax2.set_xticklabels(['','2009','','2011','','2013','','2015','','2017'], alpha = 0.3)
ax2.text(0.45, 2.45, 'OBAMA', fontsize = 18, weight='bold', color='#ffa500', transform = plt.gca().transAxes)
ax2.text(0.44, 2.34, '(2009-2017)', weight='bold', alpha=0.3, transform = plt.gca().transAxes)

### Ax3 : Trump
ax3.plot(trump['Time'], trump['rolling_mean'], color= '#00b2ee')
ax3.set_xticklabels(['','2017','','20118','','2019','','2020','','2021'], alpha = 0.3)
ax3.text(0.818, 2.45, 'TRUMP', fontsize = 18, weight='bold', color='#00b2ee', transform = plt.gca().transAxes)
ax3.text(0.808, 2.34, '(2017-2021)', weight='bold', alpha=0.3, transform = plt.gca().transAxes)

### Ax4 : Bush-Obama_Trump
ax4.plot(bush['Time'], bush['rolling_mean'], color='#6A0DAD')
ax4.plot(obama['Time'], obama['rolling_mean'], color='#ffa500')
ax4.plot(trump['Time'], trump['rolling_mean'], color='#00b2ee')

ax4.set_xticks([])

### Adding a title and a subtitle
ax1.text(-0.05, 2.8,'EURO-USD rate averaged 1.22 under the last three US presidents', fontsize=20,
        weight='bold', transform = plt.gca().transAxes)
ax1.text(-0.05,2.55, '''EURO-USD exchange rates under Georage W. Bush (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021)
''', fontsize=14, transform=plt.gca().transAxes)

plt.tight_layout()
plt.show()
```

# George W. Bush (2001-2009) - Purple Line:
There was a lot of fluctuation in the EURO-USD exchange rate when President Bush was in office. The rate climbed gradually over the following few years, beginning at parity (1.0) in 2001 and reaching a peak of little over 1.6 in 2008. During this time, the euro appreciated compared to the dollar, making the cost of purchasing euros in dollars higher. The rate fell sharply during the end of Bush's term, just around the time of the global financial crisis of 2007–2008, which had a big impact on the value of the USD.

# Barack Obama (2009-2017) - Orange Line:
The financial crisis had just ended when President Obama took office, and the currency rate had already declined from its highest point. Throughout his presidency, there were frequent swings in the currency rate between 1.2 and 1.4. Despite some ups and downs, the overall trend seems to be a mild fall, suggesting that either the euro was gradually losing value relative to the dollar throughout this time, or the dollar was gradually rising versus the euro. The exchange rate was barely above 1.0 at the end of his term.

# Donald Trump (2017-2021) - Blue Line:
The pattern of the EUR/USD exchange rate has been more stable under President Trump, with less significant fluctuations than in prior times. It demonstrated reduced volatility and a more limited range of fluctuation by staying approximately between 1.1 and 1.2. A modest increasing trend in the following years points to a potential weakening of the USD or strengthening of the euro as his term comes to an end.
